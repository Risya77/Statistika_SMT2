# ARITMATIKA DASAR

```python
# Penjumlahan
varA = 5
varB = 4
hasil_penjumlahan = varA + varB
print("Hasil Penjumlahan :", hasil_penjumlahan)

# Pengurangan
varX = 530
varY = 120
hasil_pengurangan = varX - varY
print("Hasil Pengurangan :", hasil_pengurangan)

# Perkalian
varP = 10
varQ = 5
hasil_Perkalian = varP * varQ
print("Hasil Perkalian :", hasil_Perkalian)

# Pembagian
varR = 10
varS = 2
hasil_pembagian = varR // varS
print("Hasil pembagian :", hasil_pembagian)

# Pangkat
varT = 5
varU = 2
hasil_pangkat = varT ** varU
print("Hasil pangkat :", hasil_pangkat)
```

1. Diberikan data berikut tentang tinggi badan (dalam cm) dari 10 siswa:
- 160, 165, 170, 155, 175, 162, 168, 172, 163, 158.
- Hitung rata-rata (mean), dan median dari data tersebut.
- Gunakan library numpy

```python
!pip install numpy
```

```python
import numpy as np

tb = [160, 165, 170, 155, 175, 162, 168, 172, 163, 158]
rata_rata = np.mean(tb)
print("Rata-Rata : ", rata_rata)
median = np.median(tb)
print("Median : ", median)
```

```python
# Daftar data
data = [160, 165, 170, 155, 175, 162, 168, 172, 163, 158]
total = sum(data)#menghitung total keseluruhan nilai data
mean = total / len(data)#len() = menghitung ada berapa data

print(total)
print("Mean:", mean)
```

```python
# Daftar data
data = [160, 165, 170, 155, 175, 162, 168, 172, 163, 158]

# Langkah 1: Urutkan data
data.sort()

# Langkah 2: Tentukan apakah jumlah elemen ganjil atau genap
n = len(data)
if n % 2 == 1:  # Jika ganjil
    median = data[n // 2]  # Elemen di tengah (data[5] = ambil isian di variabel data urutan ke 5)
else:  # Jika genap
    mid1 = data[n // 2 - 1]
    mid2 = data[n // 2]
    median = (mid1 + mid2) / 2  # Rata-rata dari dua elemen tengah

print("Median:", median)
```

LOGIKA KONDISIONAL (IF ELSE)

  2. Buatlah logika kondisional (if.. else) berikut:
Buat inputan angka
- Jika lebih dari 0 adalah bilangan positif
- Jika kurang dari 0 adalah bilangan negatif
- else adalah bilangan nol

```python
bilangan = int(input("Masukkan bilangan: "))

if bilangan > 0:
  print("Bilangan Positif")
elif bilangan < 0:
  print("Bilangan Negatif")
else:
  print("Bilangan Nol")
```
