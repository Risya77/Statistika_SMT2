```python
values.sort()# Langkah 1: Urutkan data
# Langkah 2: Tentukan apakah jumlah elemen ganjil atau genap
n = len(values)
if n % 2 == 1:  # Jika ganjil
    median = values[n // 2]  # Elemen di tengah (values[5] = ambil isian di variabel data urutan ke 5)
else:  # Jika genap
    mid1 = values[n // 2 - 1]
    mid2 = values[n // 2]
    median = (mid1 + mid2) / 2  # Rata-rata dari dua elemen tengah

print("Median:", median)
print (values)
```

## Measures of Variability


### **Range**

```python
import numpy

values = [13,21,21,40,48,55,72]

x = numpy.ptp(values)

print(x)
```

###Quartil & Precentiles

```python
#Quartil
import numpy

values = [13,21,21,40,42,48,55,72]

x = numpy.quantile(values, [0,0.25,0.5,0.75,1])

print(x)
```

```python
#Precentiles => mencari 65% dari data
import numpy

values = [13,21,21,40,42,48,55,72]

x = numpy.percentile(values, 65)

print(x)
```

###Interquartil Range

Interquartile range is the difference between the first and third quartiles (Q1 and Q3).

The 'middle half' of the data is between the first and third quartile.

The first quartile is the value in the data that separates the bottom 25% of values from the top 75%.

The third quartile is the value in the data that separates the bottom 75% of the values from the top 25%

```python
from scipy import stats

values = [13,21,21,40,42,48,55,72]

x = stats.iqr(values)

print(x)
```
